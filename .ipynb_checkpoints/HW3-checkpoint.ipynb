```julia
function readDat(file="bs05op.dat",start=24,stop=-3)
    f = open(file,"r")
    lines = readlines(f)
    close(f)
    s = split(lines[start]," ")
    data = [zeros(12) for i=start:length(lines)+stop]
    for i=start:length(lines)+stop
        s = filter(x -> x != "",split(lines[i]," "))
        data[i-start+1] .= parse.(Float64,s)
    end
    println(lines[start-1])
    return data
end
data = readDat()
```

```julia
μ = 0.6; mp = 1.672e-24; σ = 1e-44*(7/0.511); Rsun = 6.96e10 #cgs
τ = 0.

for i=1:length(data) #lazy 1st order integral
    Δr = i>1 ? data[i][2] - data[i-1][2] : data[1][2]
    n = data[i][4]/(μ*mp)
    τ += σ*n*Δr*Rsun
end

println("The neutrino sees τ of ≈ $(round(τ,sigdigits=3))")
println("This gives an absoprtion probability of $(round(1-exp(-τ),sigdigits=3))")
```

```julia
summaries = lines[63:188]
function getClosestTable(summaries,X,Y,Z)
    dist = zeros(length(summaries))
    for i = 1:length(summaries)
        s1 = split(summaries[i],"Solar\$    ")
        s2 = split(s1[2]," ")
        Xt,Yt,Zt = nothing, nothing, nothing
        Xt = parse(Float64,s2[1][3:end])
        Yt = parse(Float64,s2[2][3:end])
        Zt = parse(Float64,s2[3][3:end])
        dist[i] = sqrt((X-Xt)^2+(Y-Yt)^2+(Z-Zt)^2)
    end
    return findmin(dist)
end

dist,t = getClosestTable(summaries,0.75,0.25,0.)
```

```julia
function getTable(lines,tableNum,tStart = 248, tLength = 69, nextTable = 77)
    rawTable = lines[tStart+nextTable*(tableNum-1):tStart+nextTable*(tableNum-1)+tLength]
    data = zeros(19,tLength+1); logR = [-8. + 0.5*i for i=0:18]; logT = zeros(tLength+1)
    for i = 1:length(rawTable)
        s = filter(x -> x!= "",split(rawTable[i]," "))
        logT[i] = parse(Float64,s[1])
        ind = 1
        for d in s[2:end]
            n = parse(Float64,d)
            data[ind,i] = n != 9.999 ? n : 0.
            ind+=1
        end
    end
    return data,logR,logT
end

getTable(lines,1)        
```

```julia
function getClosestR(lines,X,Y,Z,logT,logR)
    summaries = lines[63:188]
    dist,t = getClosestTable(summaries,X,Y,Z)
    data,logRList,logTList = getTable(lines,t)
    Rind = findmin([(logR - logRl)^2 for logRl in logRList])[2]
    Tind = findmin([(logT - logTl)^2 for logTl in logTList])[2]
    return data[Rind,Tind]
end

getClosestR(lines,0.75,0.25,0.,log10(1.567e7),log10(0.0016))
```

```julia
function τₑ(modelData,lines)
    r,ρ,P,T,X,Y,Z,L = modelData
    τ = zeros(length(r)); crit = zeros(length(r))
    ξ = 8/15; a = 7.5657e-15; κList = zeros(length(r))
    for i=1:length(r)
        κ = getClosestR(lines,X[i],Y[i],Z[i],log10(T[i]),log10(ρ[i]/(1e-6*T[i])^3))
        κList[i] = κ
    end
    @. κModel(x,p) = p[1]*exp(p[2]*x)+p[3] #fit an exponential function
    fit = curve_fit(model,r,κ,[1.,1.,0.])
    for i=1:length(r)
        ΔP = i>1 ? (P[i]-P[i-1])/((r[i]-r[i-1])*Rsun) : P[1]/(r[1]*Rsun)
        τ[i] = κModel(r[i],fit.param)*ρ[i]*abs(P[i]/ΔP)
        Fᵣ = L[i]*Lsun/(4*π*(r[i]*Rsun)^2)
        U = a*T[i]^4
        crit[i] = ξ*c*U/Fᵣ
    end
    return τ,crit,κList,κModel,fit
end
data = readDat()
r = [d[2] for d in data]
T = [d[3] for d in data]
ρ = [d[4] for d in data]
P = [d[5] for d in data]
L = [d[6] for d in data]
X = [d[7] for d in data]
Y = [d[8] for d in data]
Z = [sum(d[9:end]) for d in data]

τ,crit,κ,κModel,fit = τₑ([r,ρ,P,T,X,Y,Z,L],lines)
p1=plot(r,τ./crit,ylims=(-0.1,3),lw=2,label="")
p1=plot!(xlabel="R [solar radii]",ylabel="τₑ/critical value",title="Ratio of τ to ξcU/Fᵣ")
p1=plot!(fontfamily="times",minorgrid=true,minorticks=true,tickdirection=:out,widen=false)
p1=plot!(r[r.>0.7],τ[r.>0.7]./crit[r.>0.7],fill=true,fillcolor=:crimson,label="Solar convection zone",fillalpha=0.5)

p2=plot(r,κ,lw=2,label="closest value in table for given parameters")
p2=plot!(r,κModel(r,fit.param),lw=2,label="exponential fit used as \"interpolation\"")
p2=plot!(fontfamily="times",minorgrid=true,minorticks=true,tickdirection=:out,widen=false)
p2=plot!(xlabel="R [solar radii]",ylabel="κ",title="Rosseland mean opacity + interpolation\nfrom N. Grevesse and A. Noels (1993)")

P = plot(p1,p2,layout=@layout([A;B]),size=(700,700),margins=2*Plots.Measures.mm)
```
