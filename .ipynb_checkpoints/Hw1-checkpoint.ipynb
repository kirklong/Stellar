```julia
g(m) = 2*m^2; E(m) = 13.6*(1-(1/m^2))
n = 1e14 #cm
k = 8.62e-5 #evK
m = 0.511e6/(3e10^2) #eV/c
h = 4.135e-15 #evs
ratio(T,n) = 1/( (1 + g(1)/g(2)*exp( (E(2)-E(1))/(k*T) )) *(1+ 2/n* (2**m*k*T/h^2)^(3/2) *exp( -(13.6-E(1))/(k*T) )) )

using Plots
T = range(2e3,stop=3e4,length=1001)
plot(T,ratio.(T,n)./1e-6,lw=2,label="n = 1e14 cm",xlabel="Temperature [K]",ylabel="(n,/n)  10",xticks=([5e3,1e4,1.5e4,2e4,2.5e4,3e4],["5,000","10,000","15,000","20,000","25,000","30,000"]),
    minorgrid=true,minorticks=true,size=(1200,700),margins=Plots.Measures.mm*5,framestyle=:box)
plot!(T,ratio.(T,n*1.75)./1e-6,label="n = 1.75e14 cm",lw=2)
```
